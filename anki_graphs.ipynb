## Vorlesung 1: Neuronale Netze

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = np.where(x < 0.5, 0, 1)
plt.plot(x, y, drawstyle='steps-post')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-8, 8, 1000)
y = 1 / (1 + np.exp(-x))
plt.plot(x, y, drawstyle='steps-post')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

print(f"sigmoid(1.5) = {sigmoid(1.5)}")
print(f"sigmoid_derivative(1.5) = {sigmoid_derivative(1.5)}")

print(f"sigmoid(4) = {sigmoid(4)}")
print(f"sigmoid_derivative(4) = {sigmoid_derivative(4)}")
```
