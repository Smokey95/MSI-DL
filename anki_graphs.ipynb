## Vorlesung 1: Neuronale Netze

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = np.where(x < 0.5, 0, 1)
plt.plot(x, y, drawstyle='steps-post')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x))
plt.plot(x, y, drawstyle='steps-post')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-5, 5, 1000)
# create a tangent hyperbolic function
y = np.tanh(x)
plt.plot(x, y, drawstyle='steps-post')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.ylim(-1.1, 1.1)
plt.grid()
plt.show()
```
